```python
import os

import pickle

import numpy as np
import pandas as pd
```

```python
DATASET_DIR = "datasets/ASSIST2009/"
DATASET_PATH = os.path.join(DATASET_DIR, "skill_builder_data.csv")
```

```python
# skill_name 결측값 제거, order_id 랑 skill_name 에서 중복 데이터 제거, order_id 기준 오름차순 정렬

df = pd.read_csv(DATASET_PATH).dropna(subset=["skill_name"])\
            .drop_duplicates(subset=["order_id", "skill_name"])\
            .sort_values(by=["order_id"])
```
